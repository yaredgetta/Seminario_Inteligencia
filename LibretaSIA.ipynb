```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
gym_data = pd.read_csv("C:/Users/alexi/OneDrive/Desktop/Seminario de IA/archive/gym_members_exercise_tracking.csv", header = 0, sep = ",")
```

```python
print(gym_data.head())
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=gym_data,
    x='BMI',
    y='Calories_Burned',
    hue='Experience_Level',
    palette='viridis'
)

plt.title('Relacion entre BMI Y calorias  QUemadas')
plt.xlabel('Indice de masa corporal (BMI)')
plt.ylabel('Calorias Quemadas')
plt.legend(title='Nivel de Experiencia')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=gym_data,
    x='BMI',
    y='Calories_Burned',
    hue='Workout_Frequency (days/week)',
    palette='viridis'
)

plt.title('Relacion entre BMI Y calorias  Quemadas')
plt.xlabel('Indice de masa corporal (BMI)')
plt.ylabel('Calorias Quemadas')
plt.legend(title='Frecuencia de Entrenamiento (dias/semana)')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=gym_data,
    x='Fat_Percentage',
    y='Calories_Burned',
    hue='Workout_Frequency (days/week)',
    palette='viridis'
)

plt.title('Relacion entre Porcentaje de grasa Y calorias  Quemadas')
plt.xlabel('Porcentaje de grasa')
plt.ylabel('Calorias Quemadas')
plt.legend(title='Frecuencia de Entrenamiento (dias/semana)')
plt.show()
```
